```python
import os
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np_file = np.load("IWCDmPMT_4pi_full_tank_e-_E0to1000MeV_unif-pos-R371-y521cm_4pi-dir_3000evts_518.npz",
                  allow_pickle=True)
time_per_event = np_file["digi_hit_time"]
charge_per_event = np_file["digi_hit_charge"]
```

```python
plt.hist(np_file["energy"])
```

# Time

```python
time_event0 = time_per_event[0]

plt.figure(figsize=(4,8))
plt.subplot(131)
plt.boxplot(time_event0)
plt.subplot(132)
plt.boxplot(time_event0)
plt.ylim([920,1100])
plt.subplot(133)
plt.boxplot(time_event0, showfliers=False)
plt.tight_layout()
```

```python
plt.hist(time_event0, bins=np.arange(920, 1100, 10))

mean = np.mean(time_event0)
plt.plot([mean, mean], [0, 1000], label='mean')

stdev = 2*np.var(time_event0)**0.5
plt.plot([mean-stdev, mean-stdev], [0, 1000], c='g', label='2*stdev')
plt.plot([mean+stdev, mean+stdev], [0, 1000], c='g')

plt.legend()
```

```python
plt.boxplot(time_per_event[:1000:51]);
```

```python
fig = plt.figure(figsize=(8,8))
fig.suptitle("Time Distribution")

plt.subplot(311)
plt.boxplot(time_per_event[:1000:51])
plt.ylabel("Time")

plt.subplot(312)
plt.boxplot(time_per_event[:1000:51])
plt.ylim([920,1100])
plt.ylabel("Time")

plt.subplot(313)
plt.boxplot(time_per_event[:1000:51], showfliers=False)
plt.ylabel("Time")
plt.xlabel("Event")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

# Charge

```python
charge_event0 = charge_per_event[0]

plt.figure(figsize=(4,8))
plt.subplot(131)
plt.boxplot(charge_event0)
plt.subplot(132)
plt.boxplot(charge_event0)
plt.ylim([-0.5,6])
plt.subplot(133)
plt.boxplot(charge_event0, showfliers=False)
plt.tight_layout()
```

```python
plt.hist(charge_event0, bins=np.arange(0, 35, 1))

mean = np.mean(charge_event0)
plt.plot([mean, mean], [0, 1000], label='mean')

stdev = 2*np.var(charge_event0)**0.5
plt.plot([mean-stdev, mean-stdev], [0, 1000], c='g', label='2*stdev')
plt.plot([mean+stdev, mean+stdev], [0, 1000], c='g')

plt.legend()
```

```python
plt.boxplot(charge_per_event[:1000:51]);
```

```python
fig = plt.figure(figsize=(8,8))
fig.suptitle("charge Distribution")

plt.subplot(311)
plt.boxplot(charge_per_event[:1000:51])
plt.ylabel("charge")

plt.subplot(312)
plt.boxplot(charge_per_event[:1000:51])
plt.ylim([-0.5,6])
plt.ylabel("charge")

plt.subplot(313)
plt.boxplot(charge_per_event[:1000:51], showfliers=False)
plt.ylabel("charge")
plt.xlabel("Event")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```
